```python
plt.figure(figsize=(9, 3))
x = ['Prix', 'Education', 'Créativité', 'Longévité', 'Développement affectif \n et social', 'Perception parentale']
y_lego = [6, 8, 8, 9, 4, 8]
y_mono = [2, 4, 3, 5, 9, 7]
plt.scatter(x, y_lego, label = 'LEGO')
plt.scatter(x, y_mono, label = 'MONOPOLY')
plt.ylim(0, 10)
plt.xticks(rotation=-20)
plt.xlabel('Attribut de valeur')
plt.ylabel("Niveau de l'offre")
plt.title('Analyse de la courbe de valeur pour le jeu LEGO et MONOPOLY')
plt.legend()
plt.savefig('Analyse_offre.pdf', bbox_inches='tight')
```
